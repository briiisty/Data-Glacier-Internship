```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
# Replace 'path_to_file' with the path to your dataset
cab_data = pd.read_csv('C:/Users/User/Desktop/DataSets-main/Cab_Data.csv')

# Create a box plot for 'Price Charged' and 'KM Travelled'
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Box plot for 'Price Charged'
ax[0].boxplot(cab_data['Price Charged'])
ax[0].set_title('Box Plot of Price Charged')
ax[0].set_ylabel('Price Charged')

# Box plot for 'KM Travelled'
ax[1].boxplot(cab_data['KM Travelled'])
ax[1].set_title('Box Plot of KM Travelled')
ax[1].set_ylabel('KM Travelled')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot for 'KM Travelled' vs 'Price Charged'
plt.figure(figsize=(10, 6))
plt.scatter(cab_data['KM Travelled'], cab_data['Price Charged'], alpha=0.5)
plt.title('Scatter Plot of KM Travelled vs Price Charged')
plt.xlabel('KM Travelled')
plt.ylabel('Price Charged')
plt.grid(True)
plt.show()
```

```python
# Calculate IQR for 'Price Charged'
Q1 = cab_data['Price Charged'].quantile(0.25)
Q3 = cab_data['Price Charged'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = cab_data[(cab_data['Price Charged'] < (Q1 - 1.5 * IQR)) | (cab_data['Price Charged'] > (Q3 + 1.5 * IQR))]

# Plot with outliers highlighted
plt.figure(figsize=(10, 6))
plt.scatter(cab_data['KM Travelled'], cab_data['Price Charged'], alpha=0.5, label='Data')
plt.scatter(outliers['KM Travelled'], outliers['Price Charged'], color='red', label='Outliers')
plt.title('Scatter Plot with Outliers Highlighted')
plt.xlabel('KM Travelled')
plt.ylabel('Price Charged')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('C:/Users/User/Desktop/DataSets-main/Cab_Data.csv')

# Ensure 'Date of Travel' is in datetime format
df['Date of Travel'] = pd.to_datetime(df['Date of Travel'], origin='1899-12-30', unit='D')

# Now you can proceed with your visualizations
# Example for seasonality in number of customers:

df['Month'] = df['Date of Travel'].dt.to_period('M')
monthly_customers = df.groupby('Month').size()

plt.figure(figsize=(12, 6))
monthly_customers.plot()
plt.title('Number of Customers per Month')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```
